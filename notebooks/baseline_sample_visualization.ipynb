```python
import torchvision.transforms as transforms
import torchvision.utils as vutils
```

```python
def visualize(sample):
    plt.figure(figsize=(10, 10))
    plt.axis('off')
    plt.imshow(np.transpose(vutils.make_grid(sample, nrow=10, 
                                             padding=2, normalize=True),
                            (1,2,0)))
    
    plt.show()
```

```python
sample1k = torch.load('../samples_ckpt200k/samples_999.pth')
```

```python
visualize(sample1k)
```

```python
sample100 = torch.load('../samples_ckpt100k/samples_0.pth')
visualize(sample100)
```
